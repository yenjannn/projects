```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### 1. 讀入資料集

```python
from google.colab import drive 
drive.mount('/content/drive') 
```

```python
import pickle
```

```python
f = open("/content/drive/MyDrive/政大應數所/論文/海報論文_Jokes/jokes/jokes_dict.pickle", 'rb')
jokesdict = pickle.load(f)
f.close()
```

```python
f = open("/content/drive/MyDrive/政大應數所/論文/海報論文_Jokes/jokes/egg_dict.pickle", 'rb')
eggdict = pickle.load(f)
f.close()
```

```python
jokesdict
```

```python
eggdict
```

```python
eggdict.keys()
```

```python
text = []
label = []
```

```python
for k in jokesdict.keys():
  text.append(k)
  label.append(jokesdict[k])
```

```python
for k in eggdict.keys():
  text.append(k)
  label.append(eggdict[k])
```

```python
print('總資料量:', len(text))
```

分割資料

```python
from sklearn.model_selection import train_test_split
```

```python
X = np.array(text)
Y = np.array(label)

X.shape, Y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=24)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

### 2. 整理資料集

```python
df = pd.DataFrame({'text':X, 'label':Y})
```

```python
df.head()
```

```python
df_train = pd.DataFrame({'text':x_train, 'label':y_train})
```

```python
df_train.head()
```

```python
df_test = pd.DataFrame({'text':x_test, 'label':y_test})
```

```python
df_test.head()
```

```python
print('訓練資料量:',len(df_train))
print('測試資料量:',len(df_test))
```

### 4. 訓練及評估

```python
model1.train_model(df)
```

### 5. 結果展示

```python
result1
```

```python
confusion_matrix1 = [[result1['tp'], result1['fn']],
            [result1['fp'], result1['tn']]]
```

```python
f, ax = plt.subplots()
ax = sns.heatmap(confusion_matrix1, cmap='YlGnBu', annot=True, fmt='d', xticklabels=False, yticklabels=False, linewidths=.5)

plt.xlabel('True')
plt.ylabel('Predicted')
plt.title('Result1')
plt.show()
```
